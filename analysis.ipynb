```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("companies.csv")
df.head()
```

```python
plt.figure(figsize=[12, 8])
sns.countplot(
    x="state",
    data=df,
)
plt.show()
```

```python
sns.barplot(x="job_openings", y="Industry", data=df)
plt.show()
```

```python
average_growth_df = (
    df[["founded", "growth_percentage"]].groupby("founded").mean().reset_index()
)
sns.lineplot(x="founded", y="growth_percentage", data=average_growth_df)
plt.show()
```

```python
sns.scatterplot(x="employees", y="estimated_revenues", data=df)
plt.show()
```

```python
sns.histplot(x="founded", data=df)
plt.show()
```
